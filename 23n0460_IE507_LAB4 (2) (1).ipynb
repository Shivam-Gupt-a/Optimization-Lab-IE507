## Plotting multiple plots

Below code snippet shows an example of having a sine and cosine curves on the same plot

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,2*np.pi,100)
f = np.sin(x)
g = np.cos(x)
plt.plot(x,f,label=r'$f(x) = \sin(x)$')
plt.plot(x,g,label=r'$g(x) = \cos(x)$')
plt.title('Sine and Cosine graph', fontsize=16)
plt.xlabel(r'$x$',fontsize=14)
plt.ylabel(r'Function of $x$', fontsize=14)
plt.legend(loc='lower left')
```

 ### ðŸ›‘ **Task 6: Read data from `scores2.csv` to perform required computation and write into a file**

Upload scores2.csv, write the above code snippet to read the file, extract the tests scores into separate lists, compute the min, max and mean of each test.

```python
import pandas as pd
import numpy as np
from numpy import mean

fin=open('scores2.csv','r+')
#reading the header file
fin.readline()
print('reading line')
# reading the rest of the line
lines=fin.readlines()
#create the list of each test and read value
test1=[]
test2=[]
test3=[]
test4=[]
for line in lines:
  mylist = line.split(',')
  test1.append(float(mylist[1]))
  test2.append(float(mylist[2]))
  test3.append(float(mylist[3]))
  test4.append(float(mylist[4]))







print('compute and store statistics in list')
maxvalues=[]
minvalues=[]
meanvalues=[]
for t in [test1,test2,test3,test4]:
    maxvalues.append(max(t))
    minvalues.append(min(t))
    meanvalues.append(mean(t))

print("Max :",maxvalues); print("Min :",minvalues); print("Mean ",meanvalues)
# creating a string to write to a file
maxstr='Max'
minstr='Min'
meanstr='Mean'
for i in range(len(maxvalues)):
  maxstr=maxstr+','+str(maxvalues[i])
  minstr=minstr+','+str(minvalues[i])
  meanstr=meanstr+','+str(meanvalues[i])
  # write to file
fin.write(maxstr+'\n'); fin.write(minstr+'\n'); fin.write(meanstr+'\n')


fin.close()
print("done")
```

## Importing and Exporting file using pandas ðŸ’¡



### Importing a csv file

Reading a CSV file in pandas is a fundamental operation that allows you to load data from a comma-separated values (CSV) file into a DataFrame. Use following command to import file named `data.csv`

`df = pd.read_csv('data.csv')  # Replace 'data.csv' with the actual file path`

Above argument uses default settings to import `csv` file, in case if you want to customize arguments while importing following below arguments:

`df = pd.read_csv('data.csv', sep=';', header=0, index_col='ID', usecols=['ID', 'Name', 'Age'], dtype={'Age': int}, na_values=['NA'])`

where,

* `sep` or `delimiter`: Specify the delimiter character. By default, it's a comma ,.
* `header`: Specify the row number (0-indexed) to be used as column names. Use header=None if there are no column names in the file.
* `index_col`: Specify the column to be used as the DataFrame's index.
* `usecols`: Specify which columns to read from the file.
* `dtype`: Specify data types for columns.
* `skiprows`: Skip a certain number of rows at the beginning of the file.
* `na_values`: List of values to be treated as NaN (missing values)
